```python
from typing import Tuple
from math import floor

import pandas as pd
```

```python
def integer_proportion_1_n(a: int, b: int) -> Tuple[int, int]:
  return (1, floor(b / a))
```

## Preparação de Dados

```python
cleaned_data = pd.read_csv("../data/academic_works.csv")
```

```python
result = (
  cleaned_data
  .groupby('db_source')['included']
  .agg(
    included_1=lambda x: (x == 1).sum(),
    included_0=lambda x: (x == 0).sum()
  )
  .reset_index()
)


result['included:excluded'] = result.apply(
  lambda row: ":".join(map(str, integer_proportion_1_n(row['included_1'], row['included_0']))),
  axis=1
)

result
```
